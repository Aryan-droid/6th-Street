# SAMPLE

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('bmh')
df = pd.read_csv('Prediction.csv')
```

```python
x=df['sku']
y=df['is_in_stock']
```

# LINE GRAPH

```python
plt.xlabel('Type of Product', fontsize=18)
plt.ylabel('Stock', fontsize=18)
plt.plot(x,y)
```

# BAR GRAPH

```python
plt.bar(x,y)
```

# SCATTER PLOT

```python
plt.scatter(x,y)
```

# PIE PLOT

```python
plt.pie(y,labels=x,radius = 1.2,autopct="%0.01f%%", shadow=True,explode=[.05,.2,.05,.2,.05,.2,.05,.2])
```

# MULTIPLE LINE GRAPH

```python
x=df['Days']
y=df['Levis']
y1=df['Lacoste']
y2=df['Tommy Hilfiger']
plt.xlabel('Day', fontsize=18)
plt.ylabel('Company', fontsize=18)
plt.plot(x,y,label = "Levis")
plt.plot(x,y1,label = "Lacoste")
plt.plot(x,y2,label = "Tommy Hilfiger")
plt.legend()
```

# SALES REPORT - ORIGINAL

```python
# Finding unique products

plt.style.use('bmh')
df = pd.read_csv('Original.csv')
unique_value = df["Item Subclass Name"].nunique()
print(unique_value)
```

# BRANDS SALABLE PRODUCTS REPORT

```python
# Accessories @ CCC Products Percentage

c = df[df['Brand']=='Accessories @ CCC']
d=c.iloc[:,9:10]
a = d.value_counts()
b= a.to_dict()
b
names = list(b.keys())
values = list(b.values())
plt.pie(values,labels=names,radius = 1.2,autopct="%0.01f%%", shadow=True)
plt.title ("Accessories @ CCC Products Percentage")
```
